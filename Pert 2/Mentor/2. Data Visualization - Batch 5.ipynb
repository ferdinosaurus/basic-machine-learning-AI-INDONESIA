### Import Library

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.style.use('seaborn')
```

### Load Data

```python
df_iris = pd.read_csv('./datasets/iris.data')
df_iris.head()
```

```python
df_iris = pd.read_csv('./datasets/iris.data', names=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class'])
df_iris.head()
```

```python
df_iris.info()
```

```python
df_iris.isnull().any()
```

```python
df_iris.describe()
```

```python
print('Total Rows:',df_iris.shape[0])
print('Total Columns:', df_iris.shape[1])
```

## Data Visualization

### Single Variable

```python
df_iris['class'].hist()
plt.show()
```

### Double Variable

```python
df_iris.plot(kind='scatter',x='sepal_length', y='sepal_width')
plt.show()
```

```python
df_iris.plot(kind='scatter', x='petal_length', y='petal_width')
plt.show()
```

### Distribution Plot

```python
sns.distplot(df_iris['sepal_length'],bins=10)
plt.show()
```

```python
sns.distplot(df_iris['petal_length'],bins=10)
plt.show()
```

```python
sns.distplot(df_iris['sepal_width'],bins=20)
```

### Box Plot

```python
sns.boxplot(data=df_iris)
plt.xlabel('Features')
plt.ylabel('Length in Cm')
plt.show()
```

```python
sns.boxplot(data=df_iris, x='class', y='petal_length')
plt.xlabel('Class')
plt.ylabel('Petal Length in Cm')
plt.show()
```

```python
sns.boxplot(data=df_iris, x='class', y='sepal_width')
plt.xlabel('Class')
plt.ylabel('Sepal Width in Cm')
plt.show()
```

### Multi Variable - Pair plot

```python
sns.pairplot(data=df_iris, hue='class')
plt.show()
```

### Multi Variable - Heat Map (Show Correlation)

```python
sns.heatmap(data=df_iris.corr(), annot=True)
plt.show()
```

```python
sns.violinplot(data=df_iris)
plt.xlabel('Features')
plt.ylabel('Length in Cm')
plt.show()
```
