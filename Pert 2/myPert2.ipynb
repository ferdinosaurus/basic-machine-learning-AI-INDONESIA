```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn
```

```python
#siapkan data
data_iris = pd.read_csv('iris.data')
data_iris.head()
```

```python
data_iris = pd.read_csv('iris.data',names=['sepal_length','sepal_width','petal_length','petal_width','class'])
data_iris.head()
```

```python
data_iris.info()
```

```python
data_iris.isnull().any()
```

```python
data_iris.describe(include='all')
```

```python
print('Total rows : ',data_iris.shape[0])
print('Total Columns',data_iris.shape[1])
```

```python
data_iris['class'].hist()
plt.show()
```

```python
data_iris.plot(kind='scatter',x='sepal_length',y='sepal_width')
plt.show()
```

```python
data_iris.plot(kind='scatter',x='petal_length',y='petal_width')
plt.show()
```

```python
seaborn.displot(data_iris['sepal_length'],bins=10)
plt.show()
```

```python
seaborn.displot(data_iris['sepal_width'],bins=10)
plt.show()
```

```python
seaborn.boxplot(data=data_iris)
plt.xlabel('Featurs')
plt.xlabel('length in cm')
plt.show()
```

```python
seaborn.boxplot(data=data_iris,x='class',y='petal_length')
plt.xlabel('class')
plt.ylabel('petal length in Cm')
plt.show()
```

```python
seaborn.boxplot(data=data_iris,x='class',y='sepal_width')
plt.xlabel('class')
plt.ylabel('petal length in Cm')
plt.show()
```

```python
seaborn.pairplot(data=data_iris,hue='class')
plt.show()
```

```python
seaborn.heatmap(data=data_iris.corr(),annot=True)
plt.show()
```

```python
seaborn.violinplot(data=data_iris)
plt.xlabel('Features')
plt.ylabel('Length in cm')
plt.show()
```
