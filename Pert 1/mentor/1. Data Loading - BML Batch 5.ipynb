### Concatenate Data

```python
df_titanic1 = df_titanic[['Survived','Pclass']]
df_titanic2 = df_titanic[['Sex','Age']]
df_titanic3 = df_titanic[['Fare']]
```

```python
df_titanic1.head()
```

```python
df_titanic3.head()
```

```python
df_titanic2.head()
```

```python
frames = [df_titanic1,df_titanic2,df_titanic3]
df_titanic_conc = pd.concat(frames,axis=1)
df_titanic_conc.head()
```

```python
df_titanic_conc.to_csv(r'./datasets/titanic_concatenate.csv')
```
